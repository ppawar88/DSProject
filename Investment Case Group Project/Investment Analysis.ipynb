# Checkpoint 2: Funding Type Analysis

```python
master_frame["raised_amount_million_usd"] = master_frame["raised_amount_usd"] / 1000000
```

```python
fundTypeGroup = master_frame.groupby(by="funding_round_type")
```

```python
avgInvestment = fundTypeGroup.mean().loc[['venture','angel','seed','private_equity']]
avgInvestment
```

```python
print('Average funding amount of venture type :', avgInvestment["raised_amount_million_usd"].loc['venture'])
print('Average funding amount of angel type :', avgInvestment["raised_amount_million_usd"].loc['angel'])
print('Average funding amount of seed type :', avgInvestment["raised_amount_million_usd"].loc['seed'])
print('Average funding amount of private equity type :', avgInvestment["raised_amount_million_usd"].loc['private_equity'])
```

```python
FT = avgInvestment[(avgInvestment["raised_amount_million_usd"] >= 5) & (avgInvestment["raised_amount_million_usd"] <= 15)].index[0]
print("Investment type suitable for spark fund (5 - 15 Million USD) : ", FT )
```

# Checkpoint 3: Country Analysis

** Loading Counties where English is official language **

Using **Contries.csv** file which is created using PDF file from <br>
http://www.emmir.org/fileadmin/user_upload/admission/Countries_where_English_is_an_official_language.pdf

```python
country = pd.read_csv("Countries.csv")
country.head()
```

```python
ventureInventment = master_frame[master_frame["funding_round_type"] == "venture"]
ventureInventment = ventureInventment.merge(country, how="left", left_on="country_code", right_on="Code")
ventureInventment.head()
```

```python
ventureInventment["Is_Eng_Official_Lang"] = ventureInventment["Code"] == ventureInventment["country_code"]
ventureInventment.head()
```

** Finding top 9 countries where most investments done **

```python
top9 = ventureInventment.groupby(by="country_code").sum()[["raised_amount_million_usd","Is_Eng_Official_Lang"]].sort_values(by="raised_amount_million_usd", ascending=False)[:9]
top9["Is_Eng_Official_Lang"] = top9["Is_Eng_Official_Lang"] > 1
top9
```

```python
top9["Rank"] = top9[["raised_amount_million_usd"]][top9["Is_Eng_Official_Lang"] == True].rank(method="max", ascending=False)
top9[top9["Is_Eng_Official_Lang"] == True]
```

```python
Country1 = top9[top9["Rank"] == 1].index[0]
Country2 = top9[top9["Rank"] == 2].index[0]
Country3 = top9[top9["Rank"] == 3].index[0]

print("Top English-speaking country with most investment : ", Country1)
print("Second English-speaking country with most investment : ", Country2)
print("Third English-speaking country with most investment : ", Country3)
```

# Checkpoint 4: Sector Analysis 1

```python
mapping = pd.read_csv("mapping.csv")
```

```python
print(mapping.shape)
mapping.head()
```

```python
mapping.dropna(subset=["category_list"], inplace=True)
print(mapping.shape)
mapping.head()
```

```python
mapping.iloc[[21]]
```

** Cleaning & Transforming Mapping data, so that can be used for deriving Sector **

Few of the category names are corrupted,<br>
Eg. category name ** A0lytics **, here the correct name should be ** Analytics **, so we have to remove **0** from such category names. 

```python
import re
mapping["category_list"] = mapping["category_list"].apply(lambda s : re.sub(r'(\w[0])*0(\s|\w+)', r'\1na\2',s).lower())
mapping.iloc[[21]]
```

```python
mapping.head()
```

Derive Main Sector name innew column for all category list

```python
#mapping["main_sector"] = mapping.apply(lambda x : (mapping.loc[x.name] == 1).argmax(), axis = 1);
sectorMapping = pd.melt(mapping, id_vars=['category_list'], var_name='main_sector')
sectorMapping = sectorMapping[sectorMapping["value"] == 1]

mapping = mapping.merge(sectorMapping[["category_list","main_sector"]], on="category_list")
mapping.head()
```

```python
master_frame["primary_sector"] = master_frame["category_list"].apply(lambda x : x.split('|')[0].lower())
```

```python
master_frame = master_frame.merge(mapping[['category_list', 'main_sector']], how="left", left_on="primary_sector", right_on="category_list", )
master_frame = master_frame.drop(["category_list_y"], axis=1)
master_frame.head()
```

```python
print("Null values for main sector : ", master_frame["main_sector"].isnull().sum() / master_frame.index.size * 100)
```

** We cannot perform sectorwise analysis on null sector values **

```python
master_frame.dropna(subset=["main_sector"], axis = 0,inplace=True)
```

# Checkpoint 5: Sector Analysis 2

Create Data Frame for **Top 3 English Speaking** countries

```python
standFilter =  (master_frame["raised_amount_million_usd"] >= 5) & (master_frame["raised_amount_million_usd"] <= 15) & (master_frame["funding_round_type"] == FT)
D1 = master_frame[(master_frame["country_code"] == Country1) & standFilter]
D2 = master_frame[(master_frame["country_code"] == Country2) & standFilter]
D3 = master_frame[(master_frame["country_code"] == Country3) & standFilter]
```

```python
print(D1.shape)
D1.head()
```

```python
print(D2.shape)
D2.head()
```

```python
print(D3.shape)
D3.head()
```

** Creating Sectorwise groups **

```python
d1SectorGroupC = D1.groupby(by="main_sector")["raised_amount_million_usd"].count()
d1SectorGroupS = D1.groupby(by="main_sector")["raised_amount_million_usd"].sum()

D1["Total_Num_Investments"] = D1["main_sector"].apply(lambda s : d1SectorGroupC[s])
D1["Total_AmountInvested"] = D1["main_sector"].apply(lambda s : d1SectorGroupS[s])
```

```python
d2SectorGroupC = D2.groupby(by="main_sector")["raised_amount_million_usd"].count()
d2SectorGroupS = D2.groupby(by="main_sector")["raised_amount_million_usd"].sum()

D2["Total_Num_Investments"] = D2["main_sector"].apply(lambda s : d2SectorGroupC[s])
D2["Total_AmountInvested"] = D2["main_sector"].apply(lambda s : d2SectorGroupS[s])
```

```python
d3SectorGroupC = D3.groupby(by="main_sector")["raised_amount_million_usd"].count()
d3SectorGroupS = D3.groupby(by="main_sector")["raised_amount_million_usd"].sum()

D3["Total_Num_Investments"] = D3["main_sector"].apply(lambda s : d3SectorGroupC[s])
D3["Total_AmountInvested"] = D3["main_sector"].apply(lambda s : d3SectorGroupS[s])
```
