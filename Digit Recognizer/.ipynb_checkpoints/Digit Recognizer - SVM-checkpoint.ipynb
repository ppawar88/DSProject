```python
Y = trainDf.pop("label")
X = trainDf
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.1, train_size=0.2, stratify=Y)
```

```python
sc = StandardScaler()
```

```python
scalledData = sc.fit_transform(X_train)
trainDf_scalled = pd.DataFrame(data=scalledData, columns=X_train.columns)
trainDf_scalled.head()
```

```python
testDf_scalled = pd.DataFrame(data=sc.transform(X_test), columns=X_test.columns)
testDf_scalled.head()
```

```python
pca = PCA(svd_solver='randomized', random_state=100)
pca.fit(trainDf_scalled)
```

```python
pca.components_.shape
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.yticks(np.arange(0,1.1,0.1))
plt.grid( linestyle='-', linewidth=0.5)
plt.show()
```

```python
from sklearn.decomposition import IncrementalPCA
final_pca = IncrementalPCA(n_components=300)
trainDf_pca = final_pca.fit_transform(trainDf_scalled)
testDf_pca = final_pca.transform(testDf_scalled)
```

```python
folds = KFold(n_splits = 5, shuffle = True, random_state = 4)
param = {'C' : [0.1, 1, 10, 100, 1000]}
svc = SVC(kernel = 'linear')
model_li = GridSearchCV(svc,param_grid=param, scoring='accuracy', cv=folds, verbose=1, return_train_score=True)
model_li.fit(trainDf_pca, Y_train)
```

```python
folds = KFold(n_splits = 3, shuffle = True, random_state = 4)
param = {'C' : [0.1, 1, 10,100],
        'gamma': [0.1, 0.01, 0.001, 0.0001]}
svc = SVC(kernel = 'poly')
model_poly = GridSearchCV(svc,param_grid=param, scoring='accuracy', cv=folds, verbose=1, return_train_score=True)
model_poly.fit(trainDf_pca, Y_train)
```

```python
folds = KFold(n_splits = 5, shuffle = True, random_state = 4)
param = {'C' : [1, 10, 100],
        'gamma': [0.1, 0.001, 0.0001]}
svc = SVC(kernel = 'rbf')
model_rbf = GridSearchCV(svc,param_grid=param, scoring='accuracy', cv=folds, verbose=1, return_train_score=True)
model_rbf.fit(trainDf_pca, Y_train)
```
